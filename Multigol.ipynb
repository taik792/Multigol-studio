```python
# @title
+11.88
1# Sostituisci TUONOME e NOMEPROGETTO con i tuoi dati reali di GitHub
import numpy as np
from scipy.stats import poisson

def ai_oracle_v21_shield():
    print("="*65)
    print("   AI ORACLE v21.0 - SHIELD & POWER ENGINE")
    print("="*65)

    try:
        # --- INPUT DATI ---
        q1 = float(input("Quota 1: "))
        qx = float(input("Quota X: "))
        q2 = float(input("Quota 2: "))
        qgg = float(input("Quota Goal: "))
        qng = float(input("Quota No Goal: "))

        print("\n[ GOL SQUADRA CASA ]")
        c_o05_q = float(input("Quota Over 0.5 Casa: "))
        c_o15_q = float(input("Quota Over 1.5 Casa: "))
        print("\n[ GOL SQUADRA OSPITE ]")
        o_o05_q = float(input("Quota Over 0.5 Ospite: "))
        o_o15_q = float(input("Quota Over 1.5 Ospite: "))

        # CALCOLO POTENZE (LAMBDA)
        l_casa = -np.log(max(0.01, 1 - (1/c_o05_q)))
        l_ospite = -np.log(max(0.01, 1 - (1/o_o05_q)))

        # Correzione pesata
        if (1/c_o15_q) < 0.45: l_casa *= 0.85
        if (1/o_o15_q) < 0.45: l_ospite *= 0.85
        l_tot = l_casa + l_ospite

        # --- 1. MERCATI TRADIZIONALI ---
        p_u25 = sum(poisson.pmf(c, l_casa) * poisson.pmf(o, l_ospite) for c in range(3) for o in range(3) if c+o <= 2) * 100
        p_o25 = 100 - p_u25
        p_gg = (1 - poisson.pmf(0, l_casa)) * (1 - poisson.pmf(0, l_ospite)) * 100
        p_ng = 100 - p_gg

        # --- 2. COMBO 1X2 + MULTIGOL TOTALE ---
        ranges_tot = [(1,3), (1,4), (2,3), (2,4), (2,5)]
        combos = []
        for mi, ma in ranges_tot:
            # Calcolo 1X + MG Totale
            p_1x = sum(poisson.pmf(c, l_casa) * poisson.pmf(o, l_ospite) for c in range(7) for o in range(7) if c >= o and mi <= (c+o) <= ma) * 100
            combos.append((p_1x, f"1X + MULTIGOL TOTALE {mi}-{ma}"))
            # Calcolo X2 + MG Totale
            p_x2 = sum(poisson.pmf(c, l_casa) * poisson.pmf(o, l_ospite) for c in range(7) for o in range(7) if o >= c and mi <= (c+o) <= ma) * 100
            combos.append((p_x2, f"X2 + MULTIGOL TOTALE {mi}-{ma}"))

        combos.sort(key=lambda x: x[0], reverse=True)

        # --- OUTPUT ---
        print("\n" + "🛡️" * 5 + " ANALISI DI MERCATO " + "🛡️" * 5)
        print(f"Under 2.5: {p_u25:.1f}%  |  Over 2.5: {p_o25:.1f}%")
        print(f"Goal:      {p_gg:.1f}%  |  No Goal:   {p_ng:.1f}%")

        print("\n" + "🎯" * 5 + " IL PRONOSTICO COMBO " + "🎯" * 5)
        print(f"CONSIGLIO: {combos[0][1]}")
        print(f"PROBABILITÀ: {combos[0][0]:.1f}%")

        # --- LOGICA DI PROTEZIONE ---
        print("\n" + "⚠️" * 5 + " ALERT DEL MOTORE " + "⚠️" * 5)
        if p_u25 > 60:
            print("ATTENZIONE: Partita da UNDER. Evita Multigol che partono da 2.")
        if p_ng > 55:
            print("ATTENZIONE: Alta probabilità NO GOAL.")
        if l_tot < 1.8:
            print("RISCHIO X FISSA: Potenza offensiva totale molto bassa.")

        print("="*65)

    except Exception as e:
        print(f"Errore: {e}")

# Avvia il programma
ai_oracle_v21_shield()
```
